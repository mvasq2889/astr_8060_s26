### Idea: 
I want to photometrically image candidates for recoiling AGN. A black hole remnant formed from a binary black hole (BBH) merger can form a recoil velocity, or "kick", due to anisotropic gravitational wave radiation during coalescence. This anisotropy stems from asymmetrical mass ratios and spin arrangements. Numerical relativity simulations predict recoil velocities can reach up to 4000km/s, which can be strong enough to displace or eject black holes out of their host galaxy entirely. Spectra can be used to identify candidates for recoiling AGN, but these candidates can be mistaken for a BBH or a spiral arm. To constrain the type of AGN, we can do one of two things: (1) track the offset of the candidate's H-beta emission line from the rest frame of their host galaxy over a long timescale or (2) try to photometrically image the host galaxy and its central AGN. (1) Tracking these AGN would involve gathering spectra over large enough timescales that we can determine whether it's a binary or not, but this can be very computationally expensive. (2) Photometric images would help us distinguish between recoiling AGN, an AGN binary, or spiral arms more quickly. If this central AGN is off-nuclear, then we can probably conclude that it is a recoiling AGN. 

 This can have major implications for the BBH populations for LISA and PTAs. We also may be able to back out information about the binary using these numerical relativity equations, which can help inform us of black hole demographics. We can also study the AGN feedback effects on galaxies.
