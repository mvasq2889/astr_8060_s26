## Measuring Mass accretion rate of Dual/Binary AGN Candidates using the Gemini Near-Infrared Imager (NIRI) 

I would like to use the Gemini North Observatory’s Near Infrared Imager (NIRI) for my project. I intend to measure the mass accretion rate (Ṁ) of dual/binary AGN candidates by estimating the mass of the black hole using broad emission lines and taking the AGN’s bolometric luminosity. These parameters allow for the calculation of an Eddington ratio and mass accretion rate. Studying the mechanics of dual/binary AGN can advance understanding of the merging processes of black holes, particularly phases leading up to gravitational wave emission. Utilizing NIRI allows for observations of AGNs that would otherwise be rendered optically invisible due to the high volume of dust and gas that their activity produces.
